```python
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
x = np.linspace(-2*np.pi,2*np.pi,1000) # returns an array from 0 to 2 with values sequetial and periodic 
```

```python
si = np.sin(x) #gives an array for each array value given 
co = np.cos(x) 
tan = np.tan(x)
ex = np.exp(x)
mask = x>0#will create a mask where x >0
validX = x[mask]#create an array where x is greater than 0 
log = np.log(validX)# then only we ca put it in log cause it only takes non zero elements 
```

```python
plt.grid(True, linestyle=':', alpha=0.6)#this will make the gridlines with spacing = 0.6 and make x and y axis bold 
plt.axhline(0, color='black', linewidth=1) # Bold line at y=0
plt.axvline(0, color='black', linewidth=1) # Bold line at x=0

plt.xlabel("X-axis")#to label or name x and y axis 
plt.ylabel("Y-axis")
plt.plot(x,si,color = "red")# plot() function takes x and y cords from two arrays and connect all the points 
plt.plot(x,co,color = "blue")
plt.plot(x,tan,color = "green")
plt.plot(x,-ex,color = "orange")
plt.plot(x,ex,color = "orange")
plt.plot([x[0],x[999]],[np.max(si),np.max(si)],color = "black",ls='--')#dashed lines representing range of sin and cos and range
plt.axhline(y = -1 ,color="black", ls="--")#A better way to whatever we did up 
plt.ylim(-2, 2)#tan has a very large range so it ruins the scale of plot minimizing sin and cos hence we limited scale to - 2 to 2 
plt.show()
```
