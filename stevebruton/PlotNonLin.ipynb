### How to plot a non linear curve in matplotlib 

#Method 1 : generate a sufficiently large number of closely-spaced x-values and their corresponding non-linear y-values using np.linspace

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp 
```

```python
# Create 100 evenly spaced points between 0 and 2*pi
x = np.linspace(0, 2 * np.pi, 100)
```

```python
# Calculate y-values for a non-linear function, e.g., y = sin(x)
y = np.sin(x)
# Or for a quadratic function, e.g., y = x**2
# y = x**2
```

```python
plt.plot(x, y, color='green') # 'green' is an optional color
plt.title("Non-Linear Function Plot (Sine Wave)")
plt.xlabel("x values")
plt.ylabel("y values")
plt.grid(True) # Optional: adds a grid for better readability
plt.show()
```

#Method 2: Solve a differential equation (ODE) and then use Matplotlib to visualize the results

```python
# 1. Define non-linear ODE: dy/dt = -y^2
def nonlinear_model(t, y):
    return -y**2

# 2. Initial conditions and time span
y0 = [2.0]
t_span = (0, 10)
t_eval = np.linspace(0, 10, 100) # Points where we want the solution

# 3. Solve the ODE
sol = solve_ivp(nonlinear_model, t_span, y0, t_eval=t_eval)

# 4. Plot using Matplotlib
plt.plot(sol.t, sol.y[0], label='Non-linear solution')
plt.xlabel('Time (t)')
plt.ylabel('y(t)')
plt.title('Non-linear ODE Solution')
plt.legend()
plt.grid(True)
plt.show()
```
